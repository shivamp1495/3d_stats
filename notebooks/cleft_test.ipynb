```python
Pk_terms_names = ['Plin', 'Pnl1', 'Pzel', 'PA', 'PW', 'Pd1', 'Pd1d1', 'Pd2', 'Pd2d2', 'Pd1d2', 'Ps2', 'Pd1s2',
                          'Pd2s2', 'Ps2s2', 'PD2', 'Pd1D2', 'k2Pk']

save_dir = '/global/u1/s/spandey/cosmosis_exp/cosmosis/y3kp-bias-model/3d_stats/bestfits/measurements/'

fig, ax = plt.subplots(1, 2, figsize=(20, 8))
cmap_step = 0.04
k=0
for j in range(len(Pk_terms_names)):
    if Pk_terms_names[j] not in [ 'PD2', 'Pd1D2', 'k2Pk']:
        xi_f = np.load(save_dir + 'xi_oneloop_cleft_bk_' + Pk_terms_names[j] + '_isregpk_False_MICE_cosmo.npz')
        color = plt.get_cmap('prism')(cmap_step*(j+1))

        if k < 7:
            ax[0].plot(xi_f['r'],np.abs(xi_f['xi'][0,:]),ls='-',label=Pk_terms_names[j],color=color)
            ax[0].plot(xi_f['r'],np.abs(-1*xi_f['xi'][0,:]),ls='--',color=color)
        else:
            ax[1].plot(xi_f['r'],np.abs(xi_f['xi'][0,:]),ls='-',label=Pk_terms_names[j],color=color)
            ax[1].plot(xi_f['r'],np.abs(-1*xi_f['xi'][0,:]),ls='--',color=color)
        k += 1
        
ax[0].set_yscale('log')
ax[0].set_xscale('log')
ax[0].set_ylabel(r'$|\xi|$', size=22)
ax[0].set_xlabel(r'$r$(Mpc/h)', size=22)
ax[0].legend(fontsize=16, frameon=False)
ax[0].tick_params(axis='both', which='major', labelsize=15)
ax[0].tick_params(axis='both', which='minor', labelsize=15)

ax[1].set_yscale('log')
ax[1].set_xscale('log')
ax[1].set_ylabel(r'$|\xi|$', size=22)
ax[1].set_xlabel(r'$r$(Mpc/h)', size=22)
ax[1].legend(fontsize=16, frameon=False)
ax[1].tick_params(axis='both', which='major', labelsize=15)
ax[1].tick_params(axis='both', which='minor', labelsize=15)

plt.tight_layout()
# plt.savefig(plot_dir + 'Cl_yy_1h_2h_comp' + '.png')
# plt.close()
            
            
```
